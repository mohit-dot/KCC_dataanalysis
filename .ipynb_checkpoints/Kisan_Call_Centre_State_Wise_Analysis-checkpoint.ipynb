## Kisan Call Centre - Data Analysis -> Generic notebook for any state for any year

Kisan Call Centres (KCCs) play a crucial role in providing valuable agricultural information and support to farmers across the country. These call centres receive numerous queries related to different seasons, crops, categories and sectors. Analyzing this data can provide valuable insights into the patterns and trends in farmers' concerns and needs. We will explore the Kisan Call Centre data and analyze the statistics of queries across key aspects such as season, crops, query types, and categories.

Data for any state for any year can be analysed by giving state name in user input.

#### Data Source : 
- Open Government Data Portal (https://data.gov.in ) . Site provides data of Kisan Call Centre from year 2007 onwards. In this notebook, I used data of year 2022 - https://data.gov.in/catalog/district-wise-and-month-wise-queries-farmers-kisan-call-centre-kcc-during-2022

 - Steps to get Request Url of catalog API : 
 1. Access the page - https://data.gov.in/catalog/district-wise-and-month-wise-queries-farmers-kisan-call-centre-kcc-during-2022
 2. Click on 'Catalog API' button. You are redirected on the swagger page.
 3. Click on the 'Generate API Key' to generate API key. This API key helps in accessing all the records of the catalog API. Default API key provides access to only 10 records.
 4. Click on 'GET'. Use provided option to get 'REQUEST URL'.

```python
# Importing libraries
from urllib.request import urlopen
import json
import pandas as pd
import numpy as np
from urllib.parse import quote

# Libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame()
limit = 2000
offset = 0
n = limit - offset
```

default api key - 579b464db66ec23bdd000001cdd3946e44ce4aad7209ff7b23ac571b

```python
# api-key mentioned here in below mentioned url is default key. Default key allows only 10 records to access. 
# I ran this notebook with my personalised API Key. You can get your personalised API key from site https://data.gov.in site. 
# Use your personalised API key in the below url

# Only 10000 records can be fetched at a time. We use Offset and Limit in loop to get all the data.  

if (total > limit):
    while (total>limit):
        url =  'https://api.data.gov.in/catalog/cd13e4de-8f42-41c1-bd42-696d6ba11a81?api-key=579b464db66ec23bdd000001cdd3946e44ce4aad7209ff7b23ac571b&format=json&filters%5BStateName%5D='+str(State_name)+'&offset='+str(offset)+'&limit='+str(limit)+''
        #print(url)
        response = urlopen(url)
        data_json = json.loads(response.read())
        data_list = data_json['records']
        df1 =  pd.DataFrame(data_list)
        df2 = df1.tail(n)
        #print(df1.shape)
        #print(df2.shape)
        df = pd.concat([df, df2])
        #print(df.shape)
        offset = limit
        limit+=2000
if (total < limit) :
    limit = total
    url =  'https://api.data.gov.in/catalog/cd13e4de-8f42-41c1-bd42-696d6ba11a81?api-key=579b464db66ec23bdd000001cdd3946e44ce4aad7209ff7b23ac571b&format=json&filters%5BStateName%5D='+str(State_name)+'&offset='+str(offset)+'&limit='+str(limit)+''
    #print(url)
    response = urlopen(url)
    data_json = json.loads(response.read())
    data_list = data_json['records']
    df1 =  pd.DataFrame(data_list)
    df2 = df1.tail(limit)
    #print(df1.shape)
    #print(df2.shape)
    df = pd.concat([df, df2])
    #print(df.shape)
       
    
```
